```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis, LinearDiscriminantAnalysis
```

#### QDA/LDA

```python
X = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])
y = np.array([0, 0, 0, 1, 1, 1])
plt.scatter(X.T[0], X.T[1], c = y, s = 100, edgecolor = 'k', linewidth = 2)
plt.title("data")
plt.show()
```

```python
model = QuadraticDiscriminantAnalysis().fit(X, y)
# model = LinearDiscriminantAnalysis().fit(X, y)
```

```python
x = [[0, 0]]
p = model.predict_proba(x)[0]

plt.subplot(2, 1, 1)
plt.scatter(X.T[0], X.T[1], c = y, s = 100, edgecolor = 'k', linewidth = 2)
plt.scatter(x[0][0], x[0][1], c = 'r', s = 100, edgecolor = 'k', marker = 'x', linewidth = 5)
plt.title("data")
plt.subplot(2, 1, 2)
plt.bar(model.classes_, p)
plt.title("conditional probability")
plt.gca().xaxis.grid(False)
plt.xticks(model.classes_, ["$ P(y = 0 | X_{test}) $", "$ P(y = 1 | X_{test}) $"])
plt.tight_layout()
plt.show()
# 가운데 있어서 각 클래스에 속할 확률 0.5 0.5
```

```python
x = [[- 0.2, - 0.1]]
p = model.predict_proba(x)[0]

plt.subplot(2, 1, 1)
plt.scatter(X.T[0], X.T[1], c = y, s = 100, edgecolor = 'k', linewidth = 2)
plt.scatter(x[0][0], x[0][1], c = 'r', s = 100, marker = 'x', edgecolor = 'k', linewidth = 5)
plt.title("data")
plt.subplot(2, 1, 2)
plt.bar(model.classes_, p)
plt.title("conditional probability")
plt.gca().xaxis.grid(False)
plt.xticks(model.classes_, ["$ P(y = 0 | X_{test}) $", "$P (y = 1 | X_{test}) $"])
plt.tight_layout()
plt.show()

# 약간 왼쪽이라 class 0에 속할 확률이 더 높음..
```

```python
x = [[0.2, 0.2]]
p = model.predict_proba(x)[0]

plt.subplot(2, 1,1)
plt.scatter(X.T[0], X.T[1], c = y, s = 100, edgecolor = 'k', linewidth = 2)
plt.scatter(x[0][0], x[0][1], c = 'r', s = 100, marker = 'x', edgecolor = 'k', linewidth = 5)
plt.title("data")
plt.subplot(2, 1, 2)
plt.bar(model.classes_, p)
plt.title("conditional probability")
plt.gca().xaxis.grid(False)
plt.xticks(model.classes_, ["$ P(y = 0 | X_{test}) $", "$ P(y = 1 | X_{test}) $"])
plt.tight_layout()
plt.show()

# 약간 오른쪽이므로 class 1에 속할 확률이 더 높음
```
